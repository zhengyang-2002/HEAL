```python
import json
import pandas as pd
```

```python
df=pd.read_json('/root/autodl-tmp/datasets/smile_conversation.jsonl',lines=True)
```

```python
final_list=[]
for index in range(len(df)):
    row=df.iloc[index]['conversation']
    temp_messageList=[{"role":"counselor","content":"你好，我是你的心理治疗师！请问有什么可以帮你？"}]+[{"role":list(item.keys())[0],"content":list(item.values())[0]} for item in row][:-1] #[{"role":"counselor","content":"你好，我是你的心理治疗师！请问有什么可以帮你？"}]+[{"role":"system","content":""}]
    temp_item=dict()
    temp_item['messages']=temp_messageList
    final_list.append(temp_item)
```

```python
final_list[:2]
```

```python
# 将数据写入 JSON 文件
with open('/root/autodl-tmp/datasets/smile_conversation_roleSwitch_shareGPT.json', 'w', encoding='utf-8') as f:
    json.dump(final_list, f, ensure_ascii=False, indent=4)

print("JSON 文件已成功创建！")
```
